```python
from supabase import create_client, Client
import pandas as pd
import os
import dotenv
dotenv.load_dotenv()
import json
```

# Firebase Init

```python
SUPABASE_URL = os.getenv('SUPABASE_URL')
SUPABASE_KEY = os.getenv('SUPABASE_KEY')
supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)
```

```python
image_folder_path = './products/images/'
df = pd.read_json('products/products.jsonl', lines=True)
df
```

```python
with open('products/products.jsonl', 'r') as file:
    products = [json.loads(line) for line in file]

# Dùng Supabase để insert dữ liệu vào bảng `products`
for product in products:
    supabase.table('products').insert(product).execute()
```

```python
# Lấy dữ liệu từ bảng 'products'
response = supabase.table('products').select('*').execute()

# Kiểm tra lỗi và lấy dữ liệu
print("Dữ liệu đã được tải lên thành công. Đây là dữ liệu trong bảng 'products':")
products = response.data
for product in products:
    print(product)  # In ra thông tin sản phẩm
```

```python
for filename in os.listdir(image_folder_path):
    if filename.endswith(".jpg") or filename.endswith(".webp"):  # Thêm webp vào điều kiện kiểm tra
        file_path = os.path.join(image_folder_path, filename)
        
        with open(file_path, 'rb') as file:
            # Upload ảnh lên storage "product-images"
            file_data = file.read()
            response = supabase.storage.from_('product-images').upload(filename, file_data)
```
